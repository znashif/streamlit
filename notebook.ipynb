## Visualization & Explanatory Analysis

### Apakah kecepatan angin (WSPM) berpengaruh terhadap penyebaran polusi udara:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=all_df, x='WSPM', y='PM2.5')
plt.title('Hubungan antara Kecepatan Angin (WSPM) dan Polusi Udara (PM2.5)')
plt.xlabel('Kecepatan Angin (m/s)')
plt.ylabel('Konsentrasi PM2.5')
plt.show()
```

### Bagaimana Hubungan antara Suhu Udara (TEMP) dan Konsentrasi Ozon (O3):

```python
# Buat scatter plot dengan regresi
plt.figure(figsize=(10, 6))
sns.regplot(data=all_df, x="TEMP", y="O3", scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})

plt.xlabel("Suhu Udara (C)")
plt.ylabel("Konsentrasi Ozon (g/m)")
plt.title("Hubungan antara Suhu Udara (TEMP) dan Konsentrasi Ozon (O3)")
plt.grid(True)
plt.show()
```

**Insight:**
- Terlihat bahwa semakin tinggi kecepatan angin, konsentrasi PM2.5 cenderung lebih rendah.
- Terlihat korelasi suhu dan ozon menunjukkan bahwa suhu udara memiliki pengaruh terhadap konsentrasi ozon, tetapi hubungannya tidak terlalu kuat dan masih dipengaruhi oleh faktor lain

## Clustering

```python
# Define bins for AQI categories (example)
bins = [0, 50, 100, 150, 200, 300, float('inf')]
labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

# Create the AQI category column
all_df['AQI_Category'] = pd.cut(all_df['PM2.5'], bins=bins, labels=labels, right=False)

# Group data by AQI category and other factors
grouped_data = all_df.groupby(['AQI_Category', 'station', 'month'])

# Perform analysis on grouped data (example: calculate the mean of other variables within each group)
grouped_means = grouped_data[['TEMP', 'WSPM', 'O3']].mean()

grouped_means
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='TEMP', y='WSPM', hue='AQI_Category', data=grouped_means.reset_index(), s=100)
plt.title('Clustering Visualization: Temperature vs. Wind Speed by AQI Category')
plt.xlabel('Temperature')
plt.ylabel('Wind Speed')
plt.legend(title='AQI Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Insight : <br>
Kategori AQI:
* Good (Biru): Umumnya berada pada kecepatan angin yang lebih tinggi.
* Moderate (Oranye) dan Unhealthy for * Sensitive Groups (Hijau): Terdistribusi di berbagai suhu dan kecepatan angin.
* Unhealthy (Merah), Very Unhealthy (Ungu), * Hazardous (Coklat): Lebih terkonsentrasi di kecepatan angin yang lebih rendah.<br>

Tren Umum:

* Kualitas udara lebih buruk (kategori merah, ungu, coklat) cenderung terjadi pada kecepatan angin yang lebih rendah.
* Kategori "Good" (biru) lebih banyak muncul pada kecepatan angin tinggi, mungkin karena dispersi polutan lebih efektif. <br>

Insight Potensial:

* Kecepatan angin tinggi dapat membantu mengurangi polusi udara, yang menjelaskan dominasi kategori "Good" di area dengan angin lebih tinggi.
* Polusi udara lebih buruk pada kecepatan angin rendah, mungkin karena stagnasi udara menyebabkan akumulasi polutan.

## Conclusion

- Apakah kecepatan angin (WSPM) berpengaruh terhadap penyebaran polusi udara?
  - Berdasarkan scatter plot yang menunjukkan hubungan antara kecepatan angin (WSPM) dan konsentrasi PM2.5, terlihat bahwa kecepatan angin berpengaruh signifikan terhadap penyebaran polusi udara.
  - Pada kecepatan angin rendah, konsentrasi PM2.5 cenderung tinggi karena polutan mengendap dan tidak tersebar dengan baik.
  - Pada kecepatan angin tinggi, konsentrasi PM2.5 menurun, menunjukkan bahwa angin membantu menyebarkan dan mendispersikan polusi udara.
Dengan demikian, kecepatan angin berperan dalam mengurangi konsentrasi polutan di udara, terutama dalam kondisi atmosfer terbuka.
- Bagaimana Hubungan antara Suhu Udara (TEMP) dan Konsentrasi Ozon (O3)?
  - Suhu udara (TEMP) memiliki korelasi positif dengan konsentrasi Ozon (O3).
  - Pada suhu yang lebih tinggi, reaksi fotokimia yang menghasilkan Ozon cenderung meningkat, sehingga konsentrasi O3 juga lebih tinggi.
  - Jika data menunjukkan tren ini dalam grafik, maka dapat disimpulkan bahwa suhu udara yang lebih tinggi mendorong pembentukan O3 lebih banyak di atmosfer.
